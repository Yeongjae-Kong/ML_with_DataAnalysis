```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import shap
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import DotProduct, WhiteKernel, Matern, ConstantKernel, ExpSineSquared, RBF, RationalQuadratic
from itertools import product
from scipy.optimize import minimize
from sklearn import preprocessing
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVR
from sklearn.model_selection import cross_validate as CV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from scipy.stats import norm

# Load the spreadsheet
file_path = 'HEA 119+5.csv'
data = pd.read_csv(file_path)
data2 = pd.read_csv('HEA 119+5.csv')

# Display the first few rows of the dataframe to understand its structure
data.head()

# Renaming columns for clarity
data.columns = ['Mn', 'Fe', 'Co', 'Ni', 'Cu', 'Mo', 'Objective']

# Splitting the data into inputs (X) and output (y)
X = data[['Mn', 'Fe', 'Co', 'Ni', 'Cu', 'Mo']].values
y = data['Objective'].values
X = (X-X.min())/(X.max()-X.min())
```

```python
# Define the kernel for Gaussian Process
# kernel = C(1.0, (1e-2, 1e7)) * RBF(10, (1e-3, 1e3))
kernel = 1.0 * Matern(5/2) 
# kernel = DotProduct() + WhiteKernel()
# kernel = Matern(5/2)
# Create Gaussian Process model
gp = GaussianProcessRegressor(kernel=kernel, alpha = 0.15, n_restarts_optimizer=1000, normalize_y = True)

# Fit the model to the data
gp.fit(X, y)

# The GP model is now trained. Next, we will perform the exploration step.
gp.kernel_  # Displaying the kernel after optimization for information
```

```python
# ----------------------------test code-------------------------------

# def generate_constrained_points(variable_index, n_points):
#     x_points = np.linspace(0, 180, n_points)
#     points = np.zeros((n_points, 6))

#     for i in range(n_points):
#         x = x_points[i]
#         points[i, variable_index] = x * 0.005

#         # 생성할 후보 점의 수
#         n_candidates = 1
#         candidates = []
#         while len(candidates) <= n_candidates:
#             # 무작위 값 생성
#             point = np.random.randint(0, 181-x, size=5)
#             # 합이 90-x와 같은지 확인
#             if point.sum() == 180-x:
#                 candidates.append(point)
        
#         # # candidates 배열 중 하나를 무작위로 선택하여 points[i]에 할당
#         selected_candidate = candidates[0]
        
#         candidate_index = 0
#         for j in range(6):
#             if j != variable_index:
#                 points[i, j] = selected_candidate[candidate_index] * 0.005
#                 candidate_index += 1
                
#     return points

# -----------------------------------------------------------
import random

def generate_constrained_points(variable_index, n_points):
    x_points = np.linspace(0, 90, n_points)
    points = np.zeros((n_points, 6))

    for i in range(n_points):
        x = x_points[i]*0.01
        points[i, variable_index] = x

        candidates = []
        # 1부터 100 사이에서 무작위로 6개의 숫자를 생성
        random_numbers = [random.randint(1, 100) for _ in range(5)]
        # 생성된 숫자들의 합 계산
        total = sum(random_numbers)
        # 합으로 나눈 결과를 계산하고 각각 소수점 두 자리까지 표현
        results = [round((0.9-x)*number / total, 2) for number in random_numbers]
        candidates.append(results)
        
        # # candidates 배열 중 하나를 무작위로 선택하여 points[i]에 할당
        selected_candidate = candidates[0]
        
        candidate_index = 0
        for j in range(6):
            if j != variable_index:
                points[i, j] = selected_candidate[candidate_index]
                candidate_index += 1
                
    return points

variable_index = 0  # Change this to select a different variable, 0 to 5
n_points = 91  # Number of points for plotting
points = generate_constrained_points(variable_index, n_points)
print("points ", points)

# Predictions and uncertainties
y_pred, sigma = gp.predict(points, return_std=True)
```

```python
def average_predictions(gp, variable_index, n_points, n_iterations=1000):
    total_y_pred = 0
    total_sigma = 0
    pointsarray = []
    for _ in range(n_iterations):
        points = generate_constrained_points(variable_index, n_points)
        pointsarray.append(points)
        y_pred, sigma = gp.predict(points, return_std=True)

        total_y_pred += y_pred
        total_sigma += sigma

    # 평균을 계산
    average_y_pred = total_y_pred / n_iterations
    average_sigma = total_sigma / n_iterations
    
    for i in range(n_iterations-1):
    # pointsarray[i]의 내용을 현재 결과 배열에 추가
        result_array = np.concatenate((pointsarray[i], pointsarray[i+1]), axis=0)

    return average_y_pred, average_sigma, result_array

def find_max_uncertainty_points(gp, sigma, candidates, n_points=10):
    # Get indices of points with highest uncertainty
    max_uncertainty_indices = np.argsort(-sigma)[:n_points]
    return candidates[max_uncertainty_indices]

def find_uncertainty_by_line(gp, n_points):
    variable_index=[1,2,3,5]
    uncertainty_list = []
    y_pred_list=[]
    sigma_list=[]
    array_list=[]
    for i in range(len(variable_index)):
        y_pred, sigma, array = average_predictions(gp, variable_index[i], n_points)
        y_pred_list.append(y_pred)
        sigma_list.append(sigma)
        array_list.append(array)
        uncertainty_list.append(100*find_max_uncertainty_points(gp, sigma, array))
    return uncertainty_list, y_pred_list, sigma_list, array_list

n_points = 91  # 생성할 점의 개수

start_time = time.time()  # 시간 check

uncertainty, y_pred, sigma, array = find_uncertainty_by_line(gp, n_points)

end_time = time.time()
total = end_time - start_time
print("prediction time :", total) 
```
